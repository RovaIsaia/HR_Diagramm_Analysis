```python
# Visualisation of star color through the histogramme
plt.figure(figsize=(14,8))
plt.bar(data1["Star color"], data1["count"], color="yellow")
# To add the values of each bar
for i,freq in enumerate(data1["count"]):
    plt.text(i, freq, str(freq), ha='center')

plt.title("Histogramme of the Star color")
plt.xlabel("Color")
plt.ylabel("Count")
plt.xticks(rotation=45)
plt.show()
```

```python
# like we have done for data1(for spectral Class)
data2 = df["Spectral Class"].value_counts().reset_index()
data2
```

```python
type(data2)
```

```python
explode = [0.1 if i == 1 else 0 for i in range(len(data2["count"]))]

fig, ax = plt.subplots(figsize=(8,6))
ax.pie(data2["count"], labels=data2["Spectral Class"], explode=explode, autopct='%1.1f%%',
       shadow=True, startangle=90)
plt.show()
```

```python
# Assuming data2 contains 'Spectral Class' and 'count' columns
total = data2["count"].sum()
percentages = (data2["count"] / total) * 100

# Use a colormap to generate different colors
colors = plt.cm.tab20.colors[:len(data2)]

fig, ax = plt.subplots(figsize=(10, 6))
bars = ax.bar(data2["Spectral Class"], percentages, color=colors)

# Add % labels on top of each bar
for bar, pct in zip(bars, percentages):
    height = bar.get_height()
    ax.text(bar.get_x() + bar.get_width() / 2, height + 0.5, f'{pct:.1f}%', 
            ha='center', va='bottom', fontsize=10)

# Add labels and title
ax.set_ylabel('Percentage (%)')
ax.set_title('Distribution of Spectral Class')
plt.xticks(rotation=45)
# Save the figure
plt.savefig("spectral_class_histogram.png")
plt.show()
```
